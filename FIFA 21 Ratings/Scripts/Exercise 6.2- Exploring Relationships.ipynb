### 3. Exploring Relationships

#### Correlation matrix:

```python
# Create a correlation matrix using pandas

fifa_21_new.corr()
```

#### Create a correlation heatmap using seaborn

```python
# Creating a subplot
f,ax = plt.subplots(figsize=(10,10))
```

```python
corr = sns.heatmap(fifa_21_new.corr(), annot = True, ax = ax) 
```

```python
corr.figure
```

#### Players ratings have a positive strong relationship of 0.58 on the wages  but value_eur have the strongest correlation of 0.84

### Scatterplots

```python
# Create a scatterplot for the "player ratings" and "wages" columns in seaborn

sns.lmplot(x = 'value_eur', y = 'wage_eur', data = fifa_21_new)
plt.xlabel("Market value(EUR)")
plt.ylabel("Wages(EUR)")
plt.title('Wages and Market value Relationship')
plt.show()
```

##### The trend line travels upwards diagonally from left to right, this shows a positive correlation between the variables, the data points are closer to the trend lines indicating a strong correlation

### Categorical Plots:

#### Creating a categorical variable that splits the "player ratings" column into categories

```python
# Use a histogram to visualize the distribution of the variables. This way, you can determine sensible categories for the price ranges. You don't want to end up with too few observations in any one of the categories.

sns.distplot(fifa_21_new['player_rating'], bins = 20)
```

```python
fifa_21_new['player_rating'].describe()
```

```python
fifa_21_new.loc[fifa_21_new['player_rating'] < 65, 'Rating Level'] = 'Low rating'
fifa_21_new.loc[(fifa_21_new['player_rating'] >= 65) & (fifa_21_new['player_rating'] < 80), 'Rating Level'] = 'Average rating'
fifa_21_new.loc[fifa_21_new['player_rating'] >= 80, 'Rating Level'] = 'High rating'
```

```python
fifa_21_new['Rating Level'].value_counts(dropna = False)
```

```python
fifa_21_new.columns
```

```python
# Create a categorical plot in seaborn using the Rating level created above

sns.set(style="ticks")
g = sns.catplot(x="wage_eur", y="player_rating", hue="Rating Level", data=fifa_21_new)
```

#### To answer the question: Do player with higher ratings receive more wages ?, 
#### With the exploratory analysis i performed i have been able to formulate the hypothesis:  
#### If player have a higher rating then they receive higher wages
#### If player have a higher value then thier wages will be higher
